```python
import sys
import os
# Adiciona o diretório raiz do projeto ao sys.path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(),'..', 'src')))

from corretoras.funcoes_bybit import busca_pnl

trades = busca_pnl(1)

ultimos = 10

print(f"Últimos {ultimos} trades, do mais recente para o mais antigo")
print(trades.head(ultimos).to_string(index=False))

print(f"\nSoma do pnl dos últimos {ultimos} trades: {round((trades['pnl'].head(ultimos)).sum(), 2)}")

total_trades = len(trades)
total_positivos = (trades['pnl'] > 0).sum()
taxa_acerto = (total_positivos / total_trades * 100) if total_trades > 0 else 0

ultimos_trades = trades.head(ultimos)
ultimos_positivos = (ultimos_trades['pnl'] > 0).sum()
taxa_acerto_ultimos = (ultimos_positivos / len(ultimos_trades) * 100) if len(ultimos_trades) > 0 else 0

print(f"Taxa de acerto para os últimos {ultimos} trades: {taxa_acerto_ultimos:.2f}%")

print(f"\nTotal de trades: {total_trades}")
print(f"Soma do pnl dos últimos {total_trades} trades: {round((trades['pnl'].head(total_trades)).sum(), 2)}")
print(f"Taxa de acerto para os últimos {total_trades} trades: {taxa_acerto:.2f}%")
```
