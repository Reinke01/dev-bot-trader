```python
import sys
import os
# Adiciona o diretório raiz do projeto ao sys.path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(),'..', 'src')))

from vectorbt_project.utils.telegram_compatibility import apply_vectorbt_telegram_patch
apply_vectorbt_telegram_patch()

from datetime import datetime
from corretoras.funcoes_bybit import carregar_dados_historicos

# Configurações
simbolo = 'BTCUSDT'
intervalo_15m = '15'
intervalo_1d = 'D'
data_inicio = '2025-01-01'
data_fim = datetime.now().strftime('%Y-%m-%d')

# Parâmetros da estratégia
ema_curta = 20
ema_longa = 45
stop = 7
rr = 2.8
periodo_rsi = 9

df_15m = carregar_dados_historicos(simbolo, intervalo_15m, [5, 15], data_inicio, data_fim)
df_15m.set_index('tempo_abertura', inplace=True)
df_15m
```

```python
import pandas_ta as ta

df_1d = carregar_dados_historicos(simbolo, intervalo_1d, [9, 21], data_inicio, data_fim)
df_1d.set_index('tempo_abertura', inplace=True)

df_1d['rsi_1d'] = ta.rsi(df_1d['fechamento'], length=periodo_rsi)
df_1d['rsi_sma_1d'] = ta.sma(df_1d['rsi_1d'], length=periodo_rsi)

df_1d.head(40)
```

```python
df_1d = df_1d.shift(1)
df_final = df_15m.join(df_1d[['rsi_1d', 'rsi_sma_1d']], on='tempo_abertura', how='left')
df_final['rsi_1d'] = df_final['rsi_1d'].ffill()
df_final['rsi_sma_1d'] = df_final['rsi_sma_1d'].ffill()

df_final.head(50)
# df_final.iloc[999]
```

```python
from entidades.estrategias_descritivas.double_ema_breakout import DoubleEmaBreakout, CondicaoEntrada, StopConfig, AlvoConfig

# Configurar estratégia
nome = "Double EMA Breakout Orders Long/Short RSI 1D"
print(f"Executando a estratégia {nome} para {simbolo} com velas de {intervalo_15m} minutos...")
estrategia = DoubleEmaBreakout(
    nome=nome,
    tipo="long_short",
    condicoes_entrada=[
        CondicaoEntrada(tipo="fechamento_acima_ema", parametros={"periodo": ema_curta}),
        CondicaoEntrada(tipo="fechamento_acima_ema", parametros={"periodo": ema_longa}),
        CondicaoEntrada(tipo="rompe_maxima_anterior", parametros={})
    ],
    stop=StopConfig(tipo="minima_das_ultimas", parametros={"quantidade": stop}),
    alvo=AlvoConfig(tipo="rr", parametros={"multiplicador": rr})
)

estrategia
```

```python
from vectorbt_project.generator_vectorbt import gerar_por_nome
import vectorbt as vbt

try:
    entries, exits, stop_price, target_price, size = gerar_por_nome("double_ema_breakout_orders_long_short_rsi1d", df_final, estrategia)

    # Criar portfólio com ordens
    order_price = entries.combine_first(exits)
    
    pf = vbt.Portfolio.from_orders(
        close=df_final['fechamento'],
        price=order_price,
        size_type='targetpercent',
        size=size,
        init_cash=1000,
        fees=0.00055,
        freq=f'{intervalo_15m}min',
        direction='both'  # Permitir posições long e short
    )
    
    # Gerar estatísticas
    stats = pf.stats()

except Exception as e:
        print(f"❌ Erro durante a execução: {str(e)}")
        raise

stats
```
